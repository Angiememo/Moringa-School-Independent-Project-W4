# Question

```python
# Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.
```

# Importing Data

```python
# We are going to import the pandas library
#
import pandas as pd

# We are going to import the numpy library
#
import numpy as np
```

# Reading the dataset given

```python
# We are going to read the dataset given through the link below:
# 'http://bit.ly/autolib_dataset'
#
df = pd.read_csv('http://bit.ly/autolib_dataset')
df
```

# Previewing the dataset

```python
# Lets preview the first 20 rows of our dataset
#
df.head(20)
```

# Accessing information

```python
# Getting to know more about the dataset by accessing its information
# 
df.info()
```

#Cleaning our datasets

Validity

```python
# We are going to drop the unneccesary information.
#
df.drop(["Displayed comment", "Scheduled at", "Status", "Subscription status"], axis = 1, inplace = True) 
df
```

```python
# Removing spaces using the str.replace() funtion
#
df['Address'] = df['Address'].str.replace("", "")
df
```

Accuracy

```python
# We will find the accuracy of the dataset.
#
df
```

Completeness

```python
# Checking if there are any missing values
#
df.isnull().values.any()
```

```python
# Dropping all attributes with any NA values
#
df_clean = df.dropna(axis=1)
print(df_clean)
```

Consistency

```python
# Removing duplicates if any.
#
df_clean_duplicate = pd.DataFrame(df_clean, columns = ['Address',	'Cars',	'Bluecar counter', 'Utilib counter',	'Utilib 1.4 counter',	'Charge Slots',	'Charging Status',	'City',	'ID',	'Kind',	'Geo point',	'Postal code',	'Public name',	'Rental status',	'Slots',	'Station type',	'year',	'month',	'day',	'hour',	'minute'])
print(df_clean_duplicate)
df_clean_non_duplicates = df_clean.drop_duplicates()
print(df_clean_non_duplicates)
```

Uniformity

```python
# Standardization by fixing messy column names
#
df_clean.columns = df_clean.columns.str.strip().str.upper().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
print(df_clean)
```

# Exporting the clean dataset

```python
# 
#
df_clean.to_csv('clean.csv')
```
